# Points by position and team

This will examine the point distribution separated by team and position. E.g. Mean/Mode points by Arsenal Midfielders or West Ham Strikers, etc..

#### Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

#### Read in "allplayers_extra.csv" and "teams.csv"

```python
data = pd.read_csv('../../data/csv/allplayers_extra.csv')

teams = pd.read_csv('../../data/csv/teams.csv')
```

#### Loop paramters

```python
# Number of teams to investigate
nTeams = 20

# Number of positions to investigate
nPositions = 4
positionStr = ['GKP', 'DEF', 'MID', 'FOR']

# player must have played at least this many minutes
minMinutes = 1

# create figure
fig, ax = plt.subplots(2,2, figsize=(16,9));

# loop through teams
for n in range(nTeams):
    
    # loop through positions
    for m in range(nPositions):
    
        c = np.logical_and(data['team']==n, data['element_type']==m)
        d = np.logical_and(c, data['minutes']>=minMinutes)
        points = data['total_points'][d]
        
        # Box-Whisker plot
        [x,y] = np.unravel_index(m, (2,2))
        ax[x,y].boxplot(points, positions=[n], widths=0.9, showmeans=True, showfliers=False)
        ax[x,y].set_title(positionStr[m])
        
# Set x-labels
for i in range(0,2):
    for j in range(0,2):
        plt.sca(ax[i,j])
        plt.xticks(range(0,nTeams), teams['short_name'], rotation=90);
        plt.yticks(np.linspace(-3, 19, 12))
        plt.grid(True)
```
