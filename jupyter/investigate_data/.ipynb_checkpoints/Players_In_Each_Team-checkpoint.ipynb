```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
elements = pd.read_csv('../../data/csv/elements.csv')
elements.head()
```

```python
players_in_team = np.zeros(20, dtype='int')

for i in range(20):
    players_in_team[i] = sum(elements.loc[:,'team']==i)
    print("Team " + str(i) + ": " + str(players_in_team[i]))
```

```python
np.median(players_in_team)
```

```python
data = pd.read_csv('../../data/csv/data01_full.csv')
data.head()
```

```python
player_played = np.zeros(90, dtype='int')
mins = np.array(range(90))
for i in range(90):
    player_played[i] = sum(data['minutes']>i)
    
plt.figure(figsize=(12,8))
plt.plot(mins,player_played)
```
