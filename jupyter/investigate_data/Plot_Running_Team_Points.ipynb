# Plot Team Points by Date

This will display each team's total points tally as a function of date

#### Import Packages

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

#### Read running total points and date time

```python
points = pd.read_csv('../../data/csv/team_running_points.csv')
points_home = pd.read_csv('../../data/csv/team_running_points_home.csv')
points_away = pd.read_csv('../../data/csv/team_running_points_away.csv')
dt = pd.read_csv('../../data/csv/team_running_datetime.csv')
```

```python
# Get column names
cols = points.columns

fig, ax = plt.subplots(4,5, figsize=(16,12))

# Convert columns to datetime
for i in range(0,20):
    [x,y]=np.unravel_index(i, (4,5))
    dt[cols[i]] = pd.to_datetime(dt[cols[i]], utc=True)
    ax[x,y].step(dt[cols[i]], points[cols[i]])
    ax[x,y].step(dt[cols[i]], points_home[cols[i]])
    ax[x,y].step(dt[cols[i]], points_away[cols[i]])
    ax[x,y].set_ylim(0,100)
    ax[x,y].set_xticks([])
    ax[x,y].set_title(cols[i])

plt.suptitle('Running Points Total', fontsize=18);
# First game of the season
#fgos = min(dt.iloc[0,:])
#lgos = max(dt.iloc[-1,:])

#dt_delta = (dt-fgos)
#dt_delta.head()
```
