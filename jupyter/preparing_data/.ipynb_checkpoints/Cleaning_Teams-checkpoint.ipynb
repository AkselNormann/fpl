# Cleaning Team Data

This will clean up the "team.json" file for efficient use and fix indexing.

#### Import Packages

```python
# Import packages
import pandas as pd
import numpy as np
import json
pd.set_option('display.max_columns', None)
```

#### Read file and convert to DataFrame

```python
# Read teams json
with open('../../data/json/teams/teams.json', 'r') as f:
    teams_json = json.load(f)
    
# Convert json to dataframe
teams = pd.DataFrame(teams_json)

print(teams.shape)
```

#### Clean up DataFrame
- Remove bloating columns

```python
# Remove bloating columns
keep_columns = ['name', 'short_name', 'strength', 'strength_attack_home', 'strength_attack_away', 'strength_defence_home', 'strength_defence_away', 'strength_overall_home', 'strength_overall_away']
teams = teams[keep_columns]
teams.head(20)
```

#### Save as csv

```python
teams.to_csv(r'../../data/csv/teams.csv', index_label=False, index=False)
```
