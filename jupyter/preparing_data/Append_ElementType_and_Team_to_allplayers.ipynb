# Append team and element_type to allplayers.csv

This will add the player's element-type (position) and team as a column in "allplayers.csv"

#### Import Packages

```python
import pandas as pd
import numpy as np
import datetime

# Set max number of columns when using head command
pd.set_option('display.max_columns', None)
```

#### Read "allplayers.csv" and "teams.csv"

```python
allplayers = pd.read_csv('../../data/csv/allplayers.csv')
elements = pd.read_csv('../../data/csv/elements.csv')
```

#### Create element-type and team columns

```python
element_type_column = elements.loc[allplayers['element'], 'element_type'].reset_index(drop=True)
team_column = elements.loc[allplayers['element'], 'team'].reset_index(drop=True)
```

#### Append to allplayers

```python
allplayers['element_type'] = element_type_column
allplayers['team'] = team_column
allplayers.head()
```

#### Save to csv

```python
allplayers.to_csv(r'../../data/csv/allplayers_extra.csv', index=False, index_label=False)
```
